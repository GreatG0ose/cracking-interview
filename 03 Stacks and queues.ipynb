```python
class Node:
    def __init__(self, value):
        self.next = None
        self.value = value

class Stack:
    def __init__(self):
        self.head = None
    
    def push(self, item):
        if self.is_empty():
            self.head = Node(item)
        else:
            new_head = Node(item)
            new_head.next = self.head
            self.head = new_head
    
    def pop(self):
        if self.is_empty():
            return
        
        self.head, self.head.next = self.head.next, None
        
    def peek(self):
        if self.is_empty():
            return None
        return self.head.value
        
    def is_empty(self):
        return self.head == None
    
class Queue:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def add(self, item):
        if self.is_empty():
            self.head, self.tail = Node(item)
        else:
            self.tail.next = Node(item)
            self.tail = self.tail.next
    
    def remove(self, item):
        if self.is_empty():
            return
        self.head, self.head.next = self.head.next, None
    
    def peek(self, item):
        if self.is_empty():
            return None
        return self.head.value
    
    def is_empty(self, item):
        return self.head == None
```

# â„–3.3 Stack of plates

```python
capacity = 5

class Stack:
    def __init__(self):
        self.storage = [0] * capacity
        self.length = 0
    
    def push(self, elem):
        if self.length == capacity:
            return False
        
        self.storage[self.length] = elem
        self.length += 1
        return True
    
    def pop(self):
        if self.length == 0:
            return None, False
        
        self.length -= 1
        elem = self.storage[self.length]
        return elem, True
    
    def peek(self):
        if self.length == 0:
            return None
        
        return self.storage[self.length - 1]
        
    
    def __str__(self):
        return f"#{self.length}\t{self.storage}"
    
class SmartStack:
    def __init__(self):
        self.stacks = []
        self.index = 0
    
    def push(self, elem):
        if len(self.stacks) < self.index + 1:
            self.stacks.append(Stack())
            
        if self.stacks[self.index].push(elem):
            return True
        else:
            self.index += 1
            return self.push(elem)
            
    def pop(self):
        elem, ok = self.stacks[self.index].pop()
        
        if ok:
            return elem, True
        
        if self.index == 0:
            return None, False
        
        self.index -= 1
        return self.pop()
    
    def popAt(self, index):
        if self.index < index:
            return None, False
        if self.index == index:
            return self.pop()
        
        return self.stacks[index].pop()
        
    
    def __str__(self):
        return f"Current: #{self.length}\n" + "\n".join([str(el) for el in self.stacks])
```

# 3.5 Sort stack

```python
def sortStack(stack):
    revStack = Stack()
    
    cap = 0
    n = stack.length
    
    for j in range(n):
        if cap == n:
            break
        maxEl = stack.peek()
        maxCount = 0

        for k in range(n - cap):
            el, ok = stack.pop()
            if not ok:
                return None

            if maxEl < el:
                maxEl = el
                maxCount = 1
            elif maxEl == el:
                maxCount += 1

            revStack.push(el)

        for k in range(n - cap):
            el, ok = revStack.pop()

            if not ok:
                return None

            if el != maxEl:
                stack.push(el)

        for k in range(maxCount):
            revStack.push(maxEl)

        cap += maxCount

    return revStack
```

```python
arr = [1, 5, 10, 2, 4]
s = Stack()

for el in arr:
    s.push(el)
    
print(sortStack(s))
```
